```python
def diss(x,y):
  p=2
  m=0
  for row in["job_title","salary"]:
     if data[row][x] == data[row][y]:
      m+=1
     return (p-m)/p

print(diss(1,1))
print(diss(1,2))
print(diss(1,3))
```
